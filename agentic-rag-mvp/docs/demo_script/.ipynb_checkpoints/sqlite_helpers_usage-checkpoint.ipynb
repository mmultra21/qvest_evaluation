# SQLite helper scripts: usage & runnable examples

This notebook documents how to run the two helper scripts that live under `agentic-rag-mvp/docs/demo_script/`:
- `sqlite_list_tables.py` — lists user tables and row counts in a SQLite DB
- `sqlite_table_details.py` — shows schema and sample rows for a given table.

You can run the commands in a Jupyter terminal (recommended) or run the runnable Python cells below which invoke the scripts using the current notebook Python interpreter. The runnable cells are guarded and will print helpful diagnostics if files are missing.

## Quick checklist before running

1. Open a Terminal in Jupyter and `cd` to the repo root: `/Users/mmultra21/Documents/qvest_evaluation`.
2. (Optional) Activate the virtualenv if you want to use the project's interpreter: `source .venv/bin/activate`.
3. Confirm the DB path exists: `agentic-rag-mvp/data/agent.db`.
4. If you prefer the command line, use the exact commands shown in the next section; or use the runnable cells to execute them from this notebook.

## Terminal commands (copy/paste into a Jupyter terminal)

Change to repo root and (optional) activate venv:

```bash
cd /Users/mmultra21/Documents/qvest_evaluation
source .venv/bin/activate  # optional
```

List tables and counts:

```bash
python3 agentic-rag-mvp/docs/demo_script/sqlite_list_tables.py agentic-rag-mvp/data/agent.db
```

Show schema + samples for a table (example `audit_logs`):

```bash
python3 agentic-rag-mvp/docs/demo_script/sqlite_table_details.py agentic-rag-mvp/data/agent.db audit_logs 10
```

````python
# Runnable: list tables using the notebook's Python interpreter (guarded).
import os, sys, subprocess
DB = os.path.join('agentic-rag-mvp','data','agent.db')
script = os.path.join('agentic-rag-mvp','docs','demo_script','sqlite_list_tables.py')
if not os.path.exists(script):
    print('Script not found at', script)
elif not os.path.exists(DB):
    print('DB not found at', DB)
else:
    python_exec = sys.executable
    print('Using Python interpreter:', python_exec)
    print('Running:', script, DB)
    r = subprocess.run([python_exec, script, DB], capture_output=True, text=True)
    print(r.stdout)
    if r.stderr:
        print('stderr:')
        print(r.stderr)
````

## Runnable: show schema + sample rows for a specified table

Change `table_name` and `sample_size` below as needed, then run the cell.

````python
# Runnable: inspect a given table (guarded)
import os, sys, subprocess
DB = os.path.join('agentic-rag-mvp','data','agent.db')
script = os.path.join('agentic-rag-mvp','docs','demo_script','sqlite_table_details.py')
table_name = 'audit_logs'  # change to audit_logs | judge_logs | hermes_raw_responses etc.
sample_size = '5'  # number of sample rows to fetch
if not os.path.exists(script):
    print('Script not found at', script)
elif not os.path.exists(DB):
    print('DB not found at', DB)
else:
    python_exec = sys.executable
    print('Using Python interpreter:', python_exec)
    cmd = [python_exec, script, DB, table_name, sample_size]
    print('Running:', ' '.join(cmd))
    r = subprocess.run(cmd, capture_output=True, text=True)
    print(r.stdout)
    if r.stderr:
        print('stderr:')
        print(r.stderr)
````

## Troubleshooting hints

- If you see `can't open file` or `No such file or directory`, confirm your current working directory (run `pwd`). The notebook kernel's cwd may differ from the Jupyter terminal. Use absolute paths if needed.
- To ensure imports and environment match the project, activate the venv in the terminal: `source .venv/bin/activate` before running the scripts in the terminal. Running from the notebook uses the kernel's interpreter (check with `sys.executable`).
- Use `ls -l agentic-rag-mvp/docs/demo_script` and `ls -l agentic-rag-mvp/data` to confirm the script and DB locations.
- If a script prints an error about SQL queries, open the DB with `sqlite3 agentic-rag-mvp/data/agent.db` and inspect tables manually.
