Inspect and run the Gradio PoC script. This notebook includes helpers to locate the script, import functions for unit testing, and example cells demonstrating how to call or launch the demo.

```python
# Helper: locate a script file from the notebook kernel (robust to cwd differences)
import os, sys, subprocess
def find_script(relpath, max_parent_depth=4):
    cwd = os.getcwd()
    cand = os.path.join(cwd, relpath)
    if os.path.exists(cand):
        return cand
    p = cwd
    for _ in range(max_parent_depth):
        p = os.path.dirname(p)
        if not p:
            break
        cand = os.path.join(p, relpath)
        if os.path.exists(cand):
            return cand
    home_docs = os.path.join(os.path.expanduser('~'), 'Documents', 'qvest_evaluation', relpath)
    if os.path.exists(home_docs):
        return home_docs
    name = os.path.basename(relpath)
    for root, dirs, files in os.walk(cwd):
        if name in files:
            return os.path.join(root, name)
    return None

REL = os.path.join('agentic-rag-mvp','scripts','run_gradio_poc.py')
```
