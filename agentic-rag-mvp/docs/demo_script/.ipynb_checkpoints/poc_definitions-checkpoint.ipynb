# POC Definitions

This notebook records concise definitions and conventions used across the librarian POC. It's intended as a lightweight glossary and runbook companion for maintainers and reviewers.


## Definition: 'Stop file'

A "stop file" is a script/module that functions as a clear hand-off or entrypoint in a workflow — a place where control stops in the pipeline (an orchestration boundary) and an explicit action is performed (for example: generate audit rows, run the judge, launch a UI). In short: the file you call (or that gets called) to perform a discrete step in the system.

### Why the term is useful

- It signals an explicit, testable boundary: you can run, inspect, or mock the stop file independently.
- Makes documentation and runbooks clearer (which file to run to trigger a behavior).
- Helps identify where to add logging, metrics, and idempotency for safe retries.

### Characteristics of a stop file

- Usually executable (has a main/CLI or a shebang and `if __name__ == '__main__'` block).
- May be invoked as a subprocess or scheduled job.
- Often imports or calls smaller libraries that do the lower-level work.
- Should be small and focused (one logical step), so it can be run and tested in isolation.
- Good to be idempotent (safe to re-run) and to persist meaningful audit/log output.

### Examples from this repo

- `agentic-rag-mvp/tools/agent_job.py` — planner entrypoint that creates audit rows. (Stop file = run this to kick off planning.)
- `agentic-rag-mvp/tools/llm_judge.py` — judge scaffold (when used as a worker or called to score candidates).
- `agentic-rag-mvp/tools/admin_ui.py` — admin helper functions / UI entrypoints used by librarians.
- `agentic-rag-mvp/tools/run_gradio_poc.py` — student-facing Gradio demo (start this to run the UI).

### How to identify stop files in a repo

Look for:
- `if __name__ == '__main__'` blocks
- Files with a shebang (#!/usr/bin/env python)
- Scripts referenced by task runners, cron entries, or runbooks
- Files invoked via subprocess.run or in Docker/compose `command` fields
- Top-level CLI frameworks (click/argparse/fire/typer)

### Practical suggestions

- Document each stop file in the repo readme or in a manual walkthrough notebook (what it does, how to run, required env vars).
- Make stop files idempotent and emit clear, structured logs and exit codes.
- Add small smoke tests that run each stop file with test fixtures.
