## Stop-files referenced by `agent_job.py`

- `agentic-rag-mvp/tools/agent_job.py` — the planner (this file).
- `agentic-rag-mvp/tools/run_gradio_poc.py` — contains the POC recommender functions such as `recommend_by_student` and `borrow_history` that the planner reuses. The planner imports this module (or loads it by filename if import fails).
- `agentic-rag-mvp/tools/llm_judge.py` — optional judge scaffold. If present, `agent_job.py` will call `judge_candidates()` after inserting audit rows.

## High-level flow (what the planner does)

1. Initialize or migrate the SQLite DB schema (`init_db()`).
2. Seed `students` table from the POC borrow history (`seed_students()`).
3. Compute planned actions for students with zero borrows (`plan_actions()`) using `poc.recommend_by_student()` to get top-k recommendations.
4. Persist actions into `audit_logs` as `pending_approval` rows (`persist_actions()`).
5. If `tools.llm_judge.judge_candidates` is available, build a small text for each inserted audit row and call the judge to score candidates (persisted by the judge to `judge_logs`).
6. Optionally apply `AUTO_APPROVE_THRESHOLD` by reading the latest judge score for each inserted audit id and marking audit rows `approved` when score >= threshold.
7. Print a short summary of inserted rows (`summarize()`).

## Key data shapes and DB schema

- `students` table: { student_id TEXT PRIMARY KEY, last_30_days_borrows INTEGER, metadata TEXT }
- `audit_logs` table: { id INTEGER PK, created_at TEXT, student_id TEXT, action_type TEXT, payload TEXT, status TEXT, notes TEXT }
- `judge_logs` table: (created by `tools/llm_judge.py`) stores audit_id, model, score, label, reason, created_at — used to determine auto-approval and for auditability.
- `hermes_raw_responses` table: optional debug table created by the judge to store raw HTTP/model responses when parsing fails.

## Important code notes (behavioral details)

- The planner uses `poc.recommend_by_student(student_id, top_k=3)` to produce top candidate tuples. The expected return shape in this POC is a list of triples (book_id, score, row_meta) — `agent_job.py` maps this to payloads of the form {book_id, score}.
- `persist_actions()` inserts one `audit_logs` row per candidate book (so 3 books => 3 rows for a student).
- If `tools.llm_judge.judge_candidates` exists, the planner builds a short `text` value for each audit row (attempting to use `label`, `title`, or other friendly fields) and passes a list of {audit_id, text} objects to the judge. The judge is responsible for writing to `judge_logs`.
- Auto-approval reads the latest judge score per audit_id and updates `audit_logs.status` to `approved` when the score >= env `AUTO_APPROVE_THRESHOLD`. If no threshold is set or no judge logs exist the planner leaves rows as `pending_approval`.
- The planner intentionally swallows judge exceptions so that judge failures don't break the scheduled job. This keeps the planner robust but can mask judge issues unless `hermes_raw_responses` are analyzed.

## Walkthrough: functions in the planner (quick map)

- `init_db(path=DB_PATH)` — creates the `students` and `audit_logs` tables using `SCHEMA`. Returns a sqlite3 connection.
- `seed_students(con)` — reads `poc.borrow_history` (from `run_gradio_poc.py`) and inserts student rows with `last_30_days_borrows=0` for demo purposes.
- `plan_actions(con)` — iterates students and for those with 0 borrows calls `poc.recommend_by_student(student_id, top_k=3)`. Transforms recommendations into `actions` list of tuples (student_id, 'recommend_email', payload).
- `persist_actions(con, actions)` — inserts one `audit_logs` row per candidate book (payload JSON). If `judge_candidates` exists it will call it with a list of `{audit_id, text}` candidates. Finally, optional auto-approval is applied using env `AUTO_APPROVE_THRESHOLD` by checking `judge_logs`.
- `summarize(con)` — prints a total count of audit rows and the last 10 rows for quick feedback.

## Operational notes and recommendations

- Auto-approval is convenient but risky. Use a conservative `AUTO_APPROVE_THRESHOLD` (example: 0.9) before enabling automatic sends to students. Keep `judge_logs` immutable for auditability.
- The planner swallows judge exceptions; ensure the judge persists raw responses (`hermes_raw_responses`) for later debugging if parsing fails. Add metric reporting to observe empty completion rates.
- For testing, run the planner repeatedly to collect more `hermes_raw_responses` samples and inspect them to determine whether the issue is model-side (empty output) or parse-side (prompt/stop tokens).
- When moving to production, migrate from SQLite to Postgres, and consider adding an approvals table that records moderator_id, timestamp, and diff of the approved payload (for better provenance).
