# Manual walkthrough: Demo flow (librarian → judge → human review → student)

This notebook walks through the end-to-end demo: what files/scripts are invoked at each step, how the LLM judge is used to auto-score candidate audit rows, how human librarians can intervene, and how students see campaign recommendations after approval. All runnable examples are guarded so the notebook is safe to run in environments missing optional services (Hermes3, sentence-transformers, Gradio).

## Goals of this walkthrough

- Explicitly call out the script / module paths invoked at each step (stop files).
- Demonstrate the automated approval pipeline using the LLM judge scaffold (`tools/llm_judge.py`).
- Show how librarians can inspect and override via `tools/admin_ui.py` helpers.
- Show how the student-facing recommendations are surfaced (Gradio / recommendation API) after approval.
- Provide CLI and Python snippets to inspect `audit_logs`, `judge_logs`, and `hermes_raw_responses` in the prototype SQLite DB (`agentic-rag-mvp/data/agent.db`).

## Prerequisites (what this demo expects)

- A Python venv in the repo root (`.venv`) with the project's runtime dependencies where available (optional).
- The prototype SQLite DB path: `agentic-rag-mvp/data/agent.db` (created by runs of the planner).
- Optional services: Hermes3 HTTP endpoint (default `http://127.0.0.1:11434`) or Hermes3 python client; sentence-transformers if you want embeddings in the sample flows; Gradio if you plan to run UIs locally. The runnable cells below are guarded if these are missing.

## Stop files (scripts/modules) referenced in the flow

1. `agentic-rag-mvp/tools/agent_job.py` — Planner that generates `recommend_email` audit rows and persists them to `audit_logs`. This is the canonical entrypoint the librarian triggers to kick off a campaign run.
2. `agentic-rag-mvp/tools/llm_judge.py` — Judge scaffold that scores pending audit rows using Hermes3 (python client → HTTP fallback) or a deterministic heuristic. It persists results to `judge_logs` and may persist unparseable Hermes responses to `hermes_raw_responses`.
3. `agentic-rag-mvp/tools/admin_ui.py` — Admin helpers / UI code used by librarians to list pending rows, view judge outputs, and perform Approve/Reject overrides.
4. `agentic-rag-mvp/tools/run_gradio_poc.py` — (Student POC): Gradio-based demo for student checkout and refetching recommendations. The student view shows campaign recommendations once audit rows are approved.
5. `agentic-rag-mvp/data/agent.db` — SQLite DB that stores `audit_logs`, `judge_logs`, and `hermes_raw_responses`.

## Step 1 — Librarian: run the planner to generate audit rows

Action: librarian (or an automated cron) runs the planner to create candidate `recommend_email` audit rows for target students. This step is implemented by calling the stop file `agentic-rag-mvp/tools/agent_job.py`.

What the planner does (high level):
- Loads or computes recommendations (TF-IDF or embeddings-based).
- Creates `audit_logs` rows with status `pending_approval` (payload includes book id, suggested text, student_id, campaign info).
- Optionally triggers the judge worker or the planner calls the judge synchronously for small pilots.

Runnable (guarded) — run the planner as a subprocess and show tail output:

```python
# Run the planner (guarded). This calls: agentic-rag-mvp/tools/agent_job.py
import subprocess, os, sys
planner = os.path.join('agentic-rag-mvp','tools','agent_job.py')
if os.path.exists(planner):
    try:
        ret = subprocess.run([os.path.join('.venv','bin','python') if os.path.exists('.venv') else sys.executable, planner], capture_output=True, text=True, check=False)
        print('planner exit code:', ret.returncode)
        print(ret.stdout[-1000:])
        if ret.stderr:
            print(ret.stderr[-1000:])
    except Exception as e:
        print('Failed to run planner:', e)
else:
    print('Planner not found at', planner)
```

What to expect after running the planner:
- New rows in `audit_logs` with status `pending_approval` (inspectable via DB queries or `admin_ui.get_pending_rows()`).
- The planner may call the judge to get an initial score; if so `judge_logs` entries will be created.
- If Hermes responses are unparseable, you'll see rows in `hermes_raw_responses` for diagnostics.

## Step 2 — Automated approval: LLM Judge scoring

Action: For each `pending_approval` audit row, the judge scaffold (`agentic-rag-mvp/tools/llm_judge.py`) is invoked. Depending on the setup, this can be called synchronously by the planner or by an asynchronous worker.

Judge responsibilities and contract:
- Input: candidate text/payload from `audit_logs`.
- Output: a JSON-like object (persisted to `judge_logs`) containing `score` (0.0-1.0) and `reason` text. The judge code tries Hermes3 first, then HTTP `/v1/completions`, then falls back to a deterministic `heuristic_score()`.
- Thresholds (example): score >= 0.85 → auto-approve; score in [0.45,0.85) → mark for review; score < 0.45 → recommend reject. These thresholds are configurable (e.g., env `AUTO_APPROVE_THRESHOLD`).

Stop file invoked: `agentic-rag-mvp/tools/llm_judge.py` — the key functions are `judge_candidates()` and the internal `_call_hermes_score()`.

Runnable (guarded) — inspect recent judge logs and show the latest scores:

```python
# Inspect recent judge_logs (guarded) and print a concise table
import sqlite3, os, json
DB = os.path.join('agentic-rag-mvp','data','agent.db')
if os.path.exists(DB):
    con = sqlite3.connect(DB)
    cur = con.cursor()
    cur.execute("SELECT id, audit_id, model, score, label, substr(reason,1,200), created_at FROM judge_logs ORDER BY created_at DESC LIMIT 20")
    rows = cur.fetchall()
    if rows:
        for r in rows:
            print(r[0], 'audit=', r[1], 'model=', r[2], 'score=', r[3], 'label=', r[4])
    else:
        print('No recent judge_logs found')
    con.close()
else:
    print('DB not found at', DB)
```

Notes:
- `judge_logs.reason` often contains either the parsed JSON from the LLM or a human-readable justification. If the LLM returned an unparseable body, the raw body may be persisted to `hermes_raw_responses` for later analysis.
- The judge scaffold includes retry and aggregation logic: it will attempt Hermes calls multiple times, parse candidates across attempts, and pick the best non-zero score when available.

Action: Librarian uses the Admin UI (or Python helpers) to list pending audit rows, inspect judge scores and reasons, and Approve or Reject items. The Admin helpers are in `agentic-rag-mvp/tools/admin_ui.py`.

Key functions / stop files referenced:
- `agentic-rag-mvp/tools/admin_ui.py::get_pending_rows()` — returns recent pending audit rows with enriched metadata (book_title if available) and judge metadata.
- `agentic-rag-mvp/tools/admin_ui.py::approve_audit_row(audit_id, moderator_id, note)` — marks an audit row approved (this function name is illustrative — inspect the real helper name in your repo).
- UI: the Gradio-based Admin sketch or any simple Flask/FastAPI admin panel can call these helpers.

What happens when a librarian approves an audit row:
- The audit row status is updated to `approved` and the approval is recorded in `audit_logs` (or a related approvals table).
- The system records who approved and any notes; `judge_logs` remains an append-only record for auditability.
- Approved recommendations are eligible to be shown to students as part of a campaign; the student-facing API/Gradio app will surface them on refresh.

## Step 4 — Student view: refresh recommendations and see campaign items

Action: After approval, the student-facing UI (Gradio demo or a production API) will surface the approved recommendations. The student POC is in `agentic-rag-mvp/tools/run_gradio_poc.py` (or via the recommendation API).

Runnable (guarded) — call the recommendation helper from the repo (sketch):


```python
# Query approved recommendations for a sample student_id from the DB (guarded)
import sqlite3, os, json
DB = os.path.join('agentic-rag-mvp','data','agent.db')
student_id = 'S1'  # replace with a real id from your seed data
if os.path.exists(DB):
    con = sqlite3.connect(DB)
    cur = con.cursor()
    # This query depends on the audit_logs payload format; adapt as needed
    cur.execute("SELECT id, payload, status, created_at FROM audit_logs WHERE status='approved' AND payload LIKE '%' || ? || '%' ORDER BY created_at DESC LIMIT 20", (student_id,))
    for row in cur.fetchall():
        print('APPROVED AUDIT:', row[0], str(row[1])[:140], row[2], row[3])
    con.close()
else:
    print('DB not found at', DB)

# Note: the demo Gradio app would simply refresh and call the recommendations API to show items; see tools/run_gradio_poc.py
```

## Step 5 — Auditability: inspecting logs and raw Hermes responses

Action: Inspect `judge_logs` and `hermes_raw_responses` to understand automatic judge behavior, failures to parse, or empty completions from Hermes. Use the sqlite3 CLI or Python as shown here.


```python
# Quick DB checks: counts and recent samples
import sqlite3, os, json
DB = os.path.join('agentic-rag-mvp','data','agent.db')
if os.path.exists(DB):
    con = sqlite3.connect(DB)
    cur = con.cursor()
    for tbl in ('audit_logs','judge_logs','hermes_raw_responses'):
        try:
            cur.execute(f"SELECT count(*) FROM {tbl}")
            print(tbl, 'count=', cur.fetchone()[0])
        except Exception as e:
            print('Table', tbl, 'not present or error:', e)
    # show a sample of raw responses (if any)
    try:
        cur.execute("SELECT id, audit_id, substr(raw_body,1,300), created_at FROM hermes_raw_responses ORDER BY created_at DESC LIMIT 10")
        for r in cur.fetchall():
            print('RAW SAMPLE', r)
    except Exception as e:
        print('No raw response samples or error:', e)
    con.close()
else:
    print('DB not found at', DB)
```

## Troubleshooting and next steps

- If you observe many zero-score judge rows with empty Hermes completions, collect `hermes_raw_responses` samples and examine the HTTP bodies. Consider prompt tightening (force FIRST-LINE JSON), increase timeouts, or switch model-serving args.
- For heavy usage, migrate `agent.db` to Postgres and secure access to judge logs with RBAC.
- Consider instrumenting the judge worker with metrics (Prometheus) to observe error/empty-completion rates and latency.
- Add end-to-end tests that exercise planner → judge → admin approval → student refresh to detect regressions early.
